```python
import feedparser
import json

def get_recent_films(username, count=5):
    url = f"https://letterboxd.com/{username}/rss/"
    feed = feedparser.parse(url)
    
    films = []
    for entry in feed.entries[:count]:
        film = {
            'title': entry.title,
            'link': entry.link,
            'id': entry.id,
            'published': entry.published,
            'watched_date': entry.letterboxd_watcheddate,
            'rewatch': entry.letterboxd_rewatch,
            'film_title': entry.letterboxd_filmtitle,
            'film_year': entry.letterboxd_filmyear,
            'rating': entry.get('letterboxd_memberrating', 'Not rated'),
            'tmdb_id': entry.tmdb_movieid,
            'summary': entry.summary,
            'author': entry.author
        }
        films.append(film)
    
    return films


username = "yerasaki"
recent = get_recent_films(username)

# josn export for website
with open('recent_films.json', 'w') as f:
    json.dump(recent, f, indent=2)

# Display
for film in recent:
    rating = film['rating']
    if rating != 'Not rated':
        rating_float = float(rating)
        full_stars = int(rating_float)
        half_star = 1 if (rating_float - full_stars) >= 0.5 else 0
        empty_stars = 5 - full_stars - half_star
        
        stars = '★' * full_stars + '½' * half_star + '☆' * empty_stars
    else:
        stars = 'No rating'
    
    print(f"{film['film_title']} ({film['film_year']})")
    print(f"Rating: {stars}")
    print(f"Watched: {film['watched_date']}\n")
```
