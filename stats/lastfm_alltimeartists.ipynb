```python
import hashlib
import requests
import json

api = 'REDACTED'
secret = 'REDACTED'
session_key = 'REDACTED'
user = 'yerasaki'

# creating signature 
sig_string = f"api_key{api}limit25methoduser.gettopartistsperiodoverallsk{session_key}user{user}{secret}"
api_sig = hashlib.md5(sig_string.encode('utf-8')).hexdigest()

# calling api
url = f"http://ws.audioscrobbler.com/2.0/?method=user.gettopartists&period=overall&user={user}&limit=25&api_key={api}&sk={session_key}&api_sig={api_sig}&format=json"

response = requests.get(url)
data = response.json()

artists = data['topartists']['artist']
for i, artist in enumerate(artists, 1):
    name = artist['name']
    playcount = artist['playcount']
    print(f"{i}. {name} - {playcount} plays")
```
